```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
raw = pd.read_csv(os.path.join('bag_covid_19_data_csv_10_December_2021/data', 'COVID19VaccSymptoms.csv'))
raw.date = pd.to_datetime(raw.date, infer_datetime_format=True)
data = raw.drop(columns={'geoRegion', 'type', 'version'})
print(data.info())
data
```

```python
print('Number of days :::::')
print(data.date.nunique())

print('\nVaccine :::::')
print(data.vaccine.value_counts(dropna=False))

print('\nSeverity :::::')
print(data.severity.value_counts(dropna=False))

print('\nAge groups :::::')
print(data.age_group.value_counts(dropna=False))

print('\nTotals number :::::')
print(data.sumTotal.nunique())
```

```python
data
```

```python
import numpy as np

sns.set(rc={"figure.dpi":150}, style='white')
plot1 = data.loc[(data['age_group']=='all') &
                (data['severity']=='all') &
                (data['vaccine']!='all')]
g=sns.lineplot(data=plot1, x='date', y='sumTotal', hue='vaccine')
plt.xticks(rotation=45, ha='right', va='baseline')
plt.xlabel('')
plt.ylabel('')
plt.legend(title=None)
sns.despine(left=True, bottom=True)


for l in g.lines:
    y = l.get_ydata()
    if len(y)>0:
        g.annotate(f'{y[-1]:.0f}', xy=(1,y[-1]), xycoords=('axes fraction', 'data'), 
                     ha='center', va='center', color=l.get_color())

g.set(yticklabels=[])

plt.show()
```

```python
dw1 = plot1[['date', 'vaccine', 'sumTotal']] #to_csv('datawrapper/plot1.csv', index=False)
dw1.pivot(index='date', columns='vaccine').to_csv('datawrapper/dw1.csv', index_label='date')
```

```python
plot1.melt(id_vars=['date', 'vaccine', 'severity', 'age_group'])
```
